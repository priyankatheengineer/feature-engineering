Label Encoding

Label Encoding is a technique used to convert categorical data into numerical form by assigning each unique category a unique integer value. Lables are assigned in alphabetical order or based on frequency of the categories.
Red: 1
Green: 2
Blue: 3

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
lbl_encoder=LabelEncoder()
```

```python
## Create a simple dataframe
df = pd.DataFrame({'color': ['red', 'blue', 'green', 'red', 'blue']})
```

```python
lbl_encoder.fit_transform(df[['color']])
```
