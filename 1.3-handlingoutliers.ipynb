```python
### Minimum, Maximum, Median, Q1, Q3, IQR
```

```python
import numpy as np 
lst_marks = [45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
minimum,Q1,median,Q3,maximum = np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0])
```

```python
minimum,Q1,median,Q3,maximum
```

```python
IQR = Q3 - Q1
print(IQR)
```

```python
## For manual computing
lower_fence = Q1 - 1.5*(IQR)
higher_fence = Q3 + 1.5*(IQR)
print(lower_fence)
print(higher_fence)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Your data
lst_marks = [45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]

# Create the boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x=lst_marks, color='skyblue')

# Add titles and labels
plt.title("Boxplot of Student Marks", fontsize=16)
plt.xlabel("Marks", fontsize=12)
plt.ylabel("Distribution", fontsize=12)  # Optional for 1D data

# Annotate min, Q1, median, Q3, max
for value, label in zip([minimum, Q1, median, Q3, maximum], ['Min', 'Q1', 'Median', 'Q3', 'Max']):
    plt.text(value, 0.02, f"{label}: {value}", 
             horizontalalignment='center', color='red', fontsize=8)

# Show the plot
plt.show()
```

```python
# data with outliers
lst_marks = [-100, -200, 45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74, 150, 170]

# Create the boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x=lst_marks, color='skyblue')

# Add titles and labels
plt.title("Boxplot of Student Marks with outliers", fontsize=16)
plt.xlabel("Marks", fontsize=12)
plt.ylabel("Distribution", fontsize=12)  # Optional for 1D data

# Annotate min, Q1, median, Q3, max
for value, label in zip([minimum, Q1, median, Q3, maximum], ['Min', 'Q1', 'Median', 'Q3', 'Max']):
    plt.text(value, 0.02, f"{label}: {value}", 
             horizontalalignment='center', color='red', fontsize=8)

# Show the plot
plt.show()
```
