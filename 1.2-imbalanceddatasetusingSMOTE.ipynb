### SMOTE (Synthetic Minority Over-sampling Technique) is a technique used to handle imbalanced datasets by creating synthetic (fake but realistic) samples for the minority class instead of simply duplicating existing ones. It is commonly used in classification problems like: Fraud detection, Credit risk modelingand Disease prediction

```python
from sklearn.datasets import make_classification
from collections import Counter
import pandas as pd

# Create synthetic imbalanced dataset
X, y = make_classification(n_samples=1000,
                           n_features=2,
                           n_redundant=0,
                           n_clusters_per_class=1,
                           weights=[0.9, 0.1],  # 90% class 0, 10% class 1
                           random_state=42)

print("Before SMOTE:", Counter(y))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(X[y==0][:,0], X[y==0][:,1], color='blue', label='Class 0', alpha=0.6)
plt.scatter(X[y==1][:,0], X[y==1][:,1], color='red', label='Class 1', alpha=0.6)

plt.title("Synthetic Imbalanced Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_resampled, y_resampled = smote.fit_resample(X, y)

print("After SMOTE:", Counter(y_resampled))
```

```python
## Plot resampled dataset
plt.scatter(X_resampled[y_resampled==0][:,0], X_resampled[y_resampled==0][:,1], color='blue', label='Class 0', alpha=0.6)
plt.scatter(X_resampled[y_resampled==1][:,0], X_resampled[y_resampled==1][:,1], color='red', label='Class 1', alpha=0.6)
plt.title("After SMOTE")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
df_resampled = pd.DataFrame(X_resampled, columns=['feature1', 'feature2'])
df_resampled['target'] = y_resampled
df_resampled
```
