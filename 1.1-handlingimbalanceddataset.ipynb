```python
import numpy as np
import pandas as pd

# Number of samples per class
n_samples = 1000
n_class_0 = int(n_samples * 0.9)
n_class_1 = n_samples - n_class_0

# Create DataFrames for each class
class_0_df = pd.DataFrame({
    'feature1': np.random.normal(0, 1, n_class_0),
    'feature2': np.random.normal(0, 1, n_class_0),
    'target': [0] * n_class_0
})

class_1_df = pd.DataFrame({
    'feature1': np.random.normal(2, 1, n_class_1),
    'feature2': np.random.normal(2, 1, n_class_1),
    'target': [1] * n_class_1
})
```

```python
class_0_df
```

```python
class_1_df
```

```python
# Combine into a single dataset
data = pd.concat([class_0_df, class_1_df], ignore_index=True)
```

```python
data
```

```python
from sklearn.utils import resample

# Separate majority and minority
class_0_data = data[data['target'] == 0]
class_1_data = data[data['target'] == 1]

# Oversample minority class to match majority
class_1_oversampled = resample(class_1_data, 
                               replace=True,       # sample with replacement
                               n_samples=len(class_0_data), # match majority
                               random_state=123)

# Combine back into balanced dataset
balanced_data = pd.concat([class_0_data, class_1_oversampled], ignore_index=True)

# Shuffle the dataset
balanced_data = balanced_data.sample(frac=1, random_state=123).reset_index(drop=True)

# Check counts
print(balanced_data['target'].value_counts())
```

```python
# Separate majority and minority
class_0_data = data[data['target'] == 0]
class_1_data = data[data['target'] == 1]

# Downsample majority class to match minority
class_0_downsampled = resample(class_0_data,
                               replace=False,      # no replacement, just pick a subset
                               n_samples=len(class_1_data),  # match minority size
                               random_state=123)

# Combine downsampled majority with minority class
balanced_data = pd.concat([class_0_downsampled, class_1_data], ignore_index=True)

# Shuffle the dataset
balanced_data = balanced_data.sample(frac=1, random_state=123).reset_index(drop=True)

# Check counts
print(balanced_data['target'].value_counts())
```
