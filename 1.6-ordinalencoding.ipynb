Ordinal Encoding is a technique used to convert categorical data with a meaningful order into numerical values. Unlike nominal data, ordinal data has a ranking or order, so assigning numbers makes sense.
Example:
1. High School: 1
2. Intermediate: 2
3. Graduation: 3
4. Post Graduation: 4


```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd
```

```python
df = pd.DataFrame({'size': ['small','medium', 'large', 'medium', 'small', 'large']})
```

```python
df
```

```python
## create instance of ordinal encoder
encoder = OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```
